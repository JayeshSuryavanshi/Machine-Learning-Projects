```python
# Missing values
dataframe2.isna().sum()
dataframe2.dropna(axis = 0, inplace = True)
```

```python
unique_countries = np.unique(dataframe2['Country'])
res = []
for x in unique_countries:
    res.append(dataframe2[dataframe2['Country'] == x]['AverageTemperature'].mean())

temperatures = [x for x in res]
final = []
for temp in temperatures:
  final.append(temp)

print(len(final))

# Reference: https://plotly.com/python/choropleth-maps/

# fig = px.choropleth(df, geojson=geojson, color="Bergeron",
#                     locations="district", featureidkey="properties.district",
#                     projection="mercator"
#                    )
# fig.update_geos(fitbounds="locations", visible=False)
# fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
# fig.show()

mapping = graph_objects.Figure(data = graph_objects.Choropleth(
    locations = unique_countries, z = res, locationmode = 'country names', text = unique_countries, colorscale = 'balance',
    colorbar_title = "Average Temperature",
))

mapping.show()    
```

```python
# dataframe3 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/GlobalLandTemperaturesByCountry.csv')
unique_countries = dataframe3['Country'].unique()
print(len(unique_countries))
pair_list = []

for country in unique_countries:
    match = dataframe3['Country'] == country
    present = dataframe3[match]['AverageTemperature']
    pair = (present.max(), present.min())
    pair_list.append(pair)
    
res_pair_list = []
final = []
differences = []

for i in range(len(pair_list)):
    if not np.isnan(pair_list[i][0]):
        res_pair_list.append(pair_list[i])
        final.append(unique_countries[i])

for x in res_pair_list:
    differences.append(x[0] - x[1])
    
differences, final = (list(x) for x in zip(*sorted(zip(differences, final), key=lambda pair: pair[0], reverse=True)))

f, ax = plt.subplots(figsize=(10,10))
sns.barplot(x=differences[:25], y=final[:25], palette=sns.color_palette("coolwarm", 30), ax=ax)

labels = ax.set(ylabel="", xlabel="The difference in temperatures", title="Countries with the highest temperature differences")
```
